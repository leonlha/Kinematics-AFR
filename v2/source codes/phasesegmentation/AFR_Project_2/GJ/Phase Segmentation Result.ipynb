```python
import matplotlib.pyplot as plt
import numpy as np

# Read data from the file
video_frames = {}

# Open the file in read mode
with open("train.txt", "r") as file:
    # Read all lines from the file
    lines = file.readlines()

# Process each line from the file
for line in lines:
    # Remove any leading or trailing whitespace characters
    line = line.strip()
    
    # Split the line by "---" to separate file name and metadata
    parts = line.split("---")
    
    # Extract the file name and frame number
    file_name = parts[0]  # Get the first part before "---"

    # Extract the class name
    class_name = parts[1].strip()  # Get the class name from the second part
    
    # Extract the frame number from the file name
    frame_part = file_name.split("Frame")[1]  # Extract part after "Frame"
    frame_number_str = frame_part.split(".")[0]  # Extract part before ".jpg"
    frame_number = int(frame_number_str)  # Convert frame number string to integer
    
    # Get the video name (prefix of the file name)
    video_name = file_name.split("Frame")[0]
    
    # Update dictionary with video name, frame number, and class name
    if video_name not in video_frames:
        video_frames[video_name] = []
    video_frames[video_name].append((frame_number, class_name))

# # Print the length of the list associated with each video name
# for video_name in video_frames:
#     print(f"{video_name}: {len(video_frames[video_name])}")

# # Print the video frames dictionary
# for video_name, frames in video_frames.items():
#     print(f"Video: {video_name}")
#     for frame_number, class_name in frames:
#         print(f"Frame {frame_number}: Class {class_name}")

first_video_name = list(video_frames.keys())[0]
print('first_video_name', first_video_name)

print(len(video_frames[first_video_name]))    
print((video_frames[first_video_name]))

# Create a new sorted list by the first element of each tuple
sorted_frames = sorted(video_frames[first_video_name], key=lambda x: x[0])

# Print the sorted list
print("Sorted list:", sorted_frames)
```
